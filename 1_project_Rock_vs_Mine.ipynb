```python
# подключение библиотек
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Сбор и обработка данных

```python
# загрузка наборв данных при помощт pandas
sonar_data = pd.read_csv('sonar data.csv', header=None)
```

```python
sonar_data.head() # Показать первые 5 столбцов Датасета
```

```python
# Количество строк и столбцов
sonar_data.shape # 208 строк и 61 столбец
```

```python
# Статистические определения для данных
sonar_data.describe()
# Количество значений
# среднее значение столбца
# отклонение
# среднее значение
# минмальное отклонение
# процентные соотношения
# максимальное значение
```

```python
sonar_data[60].value_counts() # Количество мин и камней
```

```python
sonar_data.groupby(60).mean()
```

```python
# Разделяем данные и метки
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

Данные тренировок и тестов

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print( X_train,  Y_train)
```

Обучение модели
Модель логистической регрессии

```python
model = LogisticRegression()
```

```python
# обучаем модель логистической регрессии
model.fit(X_train, Y_train)
```

Оценка модели

```python
# точность по данным обучения
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Точность для тренировочных данных: ', training_data_accuracy)
```

```python
# точность по тестовым данным обучения
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Точность для тестовых данных: ', test_data_accuracy)
```
